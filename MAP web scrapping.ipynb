```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from datetime import datetime
import time

# SETUP DRIVER
options = Options()
options.add_argument("--start-maximized")
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
    "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36"
)
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
wait = WebDriverWait(driver, 25)

# PRODUCT LIST
products = [
    {
        "Category": "Laptop",
        "Brand": "HP",
        "Amazon": "https://www.amazon.in/dp/B0FM4FNVB8",
        "Flipkart": "https://www.flipkart.com/hp-15-intel-core-i5-13th-gen-1335u-16-gb-512-gb-ssd-windows-11-home-15-fd0576tu-15-fd0467tu-thin-light-laptop/p/itm3d552029526ba?pid=COMH8QXWHJRQKR6P&lid=LSTCOMH8QXWHJRQKR6PW3FWJH&marketplace=FLIPKART&q=HP+Laptop+15%2C+Intel+Core+i5-1335U-13th+Gen&store=6bo%2Fb5g&srno=s_1_1&otracker=search&otracker1=search&fm=organic&iid=en_dxVNFAqiej26qoYFscd8b9UC0twF1KM-4C5vuEKtHt4lNVVuxt0yLqz12PDRm_XM9cVpoYRpUZV1iH1OtY8wXPUFjCTyOHoHZs-Z5_PS_w0%3D&ppt=None&ppn=None&ssid=oz0gefzda80000001771396662916&qH=b88374192f874d8c&ov_redirect=true"
    },
    {
        "Category": "Ink",
        "Brand": "HP",
        "Amazon": "https://www.amazon.in/HP-680-Tri-color-Advantage-Cartridge/dp/B014AL6Y7W/ref=sr_1_1_sspa?crid=1R3BQGRNN8FOI&dib=eyJ2IjoiMSJ9.f8VJWH19I-aYrt7GrYiP7umYrtBRiwQuenodtOsH-LwFkcNYXKWfvN9P8-rrmj8K7QWsqXnrN4sLkdJi4pIKC04Y1J6NccW4Z_6pwfmMeBS2DTAZP38JdIAE79Fg4N9T6fi9LTxkcimfsc70jiDM2wKfnRfUIpA8Usc-O94I0LWTF9E9uHVEnSKl8EW2rVwX5b3m8c8n3f3QTEhSe9AqwWoR5EV1y80paqbTB5TiiSk.8lkE_TWJnYuPh5Fc2ulZSFWp1emkUchRlD1GdkN1ab4&dib_tag=se&keywords=ink%2Bfor%2Bprinter%2Bhp&qid=1771399205&sprefix=Ink%2Bfor%2Bprinter%2Caps%2C506&sr=8-1-spons&aref=KxvqY0SHlC&sp_csd=d2lkZ2V0TmFtZT1zcF9hdGY&th=1",
        "Flipkart": "https://www.flipkart.com/hp-680-tri-color-ink-cartridge/p/itmc9bd71d98f127?pid=INKECWFBMS4VGT8J&lid=LSTINKECWFBMS4VGT8JO1ZO09&marketplace=FLIPKART&q=HP+680+Tri-color&store=6bo%2Ftia%2Fffn%2Fadt&srno=s_1_4&otracker=search&otracker1=search&fm=Search&iid=651f1e40-a480-49a4-9318-791a62df1bec.INKECWFBMS4VGT8J.SEARCH&ppt=sp&ppn=sp&ssid=ygnn33xz400000001771399223067&qH=ad47b67f2875325c&ov_redirect=true&ov_redirect=true"
    },
    {
        "Category": "Printer",
        "Brand": "HP",
        "Amazon": "https://www.amazon.in/HP-Smart-Colour-Printer-Included/dp/B0BN1XT6TF/ref=sr_1_1_sspa?crid=TH7ZKN8545XC&dib=eyJ2IjoiMSJ9.Yfwo959MJn1QKmxepyJZg67oVEh5OSnwntMD40DXRkDeUus6lcYn-aAqQh40PMRg8D0et2LrNAib0X-N6OeyZC39P2ca7ca4nOliVmm9R9GDpGo9enu5H9M5ErMtkKa0rU3XS8-D_uhidHvrkXgQzAjsSVtusQ9KqwJVpCX5csYIPWDnfOAawHfncspxdyOy-TAXvHTGL28qD8Ht-V66ntJzGcUiOTJ7sdLoQtnoKDo._YdEeBf2G2Rq3jZIF9vs2sC88yl2EUYiZzDIKBTkJzE&dib_tag=se&keywords=printer%2Bhp&qid=1771399329&sprefix=printer%2Bh%2Caps%2C407&sr=8-1-spons&aref=yO4lJQnZsV&sp_csd=d2lkZ2V0TmFtZT1zcF9hdGY&th=1",
        "Flipkart": "https://www.flipkart.com/hp-smart-all-one-589-multi-function-wifi-color-ink-tank-printer-print-scan-copy-up-4000-black-6000-pages-box/p/itmd0ee2f16df471?pid=PRNGKZB6NY9YMRJN&lid=LSTPRNGKZB6NY9YMRJNGYQULQ&marketplace=FLIPKART&q=HP+Smart+Tank+589+All-in-One+WiFi+Colour+Printer&store=6bo%2Ftia%2Fffn%2Ft64&spotlightTagId=default_BestsellerId_6bo%2Ftia%2Fffn%2Ft64&srno=s_1_1&otracker=search&otracker1=search&fm=Search&iid=96e35b1e-d47e-4a92-a18c-100ae686fb89.PRNGKZB6NY9YMRJN.SEARCH&ppt=sp&ppn=sp&ssid=d4sxz9jhkw0000001771399345636&qH=77968e040c16f10e&ov_redirect=true&ov_redirect=true"
    }
]

# AMAZON SCRAPER
def scrape_amazon(url):
    driver.get(url)
    sku = url.split("/dp/")[1].split("/")[0].split("?")[0] if "/dp/" in url else "Unknown"
    price = "N/A"
    seller = "Unknown"
    try:
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.ID, "productTitle"))
        )
    except:
        print(f"  ⚠️  [Amazon] Page may be a CAPTCHA or redirect — saving debug screenshot.")
        driver.save_screenshot(f"DEBUG_Amazon_{sku}.png")
        return None

    driver.execute_script("window.scrollBy(0, 300);")
    time.sleep(1)
    driver.execute_script("window.scrollBy(0, 300);")
    time.sleep(2)

    # --- Price ---
    try:
        price_elem = driver.find_element(By.CSS_SELECTOR, "span.a-price span.a-offscreen")
        price = price_elem.get_attribute("textContent").strip()
    except:
        try:
            price = driver.find_element(By.ID, "priceblock_ourprice").text.strip()
        except:
            price = "N/A"
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "#tabular-buybox .tabular-buybox-text"))
        )
        labels = driver.find_elements(By.CSS_SELECTOR, "#tabular-buybox .tabular-buybox-text")
        values = driver.find_elements(By.CSS_SELECTOR, "#tabular-buybox .tabular-buybox-text-message")
        for label_el, value_el in zip(labels, values):
            if "Sold by" in label_el.text.strip():
                try:
                    seller = value_el.find_element(By.TAG_NAME, "a").text.strip()
                except:
                    seller = value_el.text.strip()
                if seller:
                    print(f"  ✅ [Amazon] Method 1 found seller: {seller}")
                    break
    except:
        pass 
    if not seller or seller == "Unknown":
        try:
            sold_by_label = driver.find_element(
                By.XPATH,
                "//span[contains(@class,'tabular-buybox-text') and contains(text(),'Sold by')]"
            )
            value_span = driver.find_element(
                By.XPATH,
                "//span[contains(@class,'tabular-buybox-text') and contains(text(),'Sold by')]"
                "/following::span[contains(@class,'tabular-buybox-text-message')][1]"
            )
            seller = value_span.text.strip()
            if seller:
                print(f"  ✅ [Amazon] Method 2 found seller: {seller}")
        except:
            pass

    if not seller or seller == "Unknown":
        try:
            mi = driver.find_element(By.ID, "merchant-info")
            try:
                seller = mi.find_element(By.TAG_NAME, "a").text.strip()
                print(f"  ✅ [Amazon] Method 3 (merchant-info anchor) found seller: {seller}")
            except:
                full = mi.text.strip()
                if "Sold by" in full:
                    raw = full.split("Sold by")[1]
                    for d in ["\n", "Ships from", "."]:
                        raw = raw.split(d)[0]
                    seller = raw.strip()
                elif "Ships from and sold by" in full:
                    seller = full.split("Ships from and sold by")[1].split(".")[0].strip()
                if seller:
                    print(f"  ✅ [Amazon] Method 3 (merchant-info text) found seller: {seller}")
        except:
            pass

    if not seller or seller == "Unknown":
        try:
            body_text = driver.find_element(By.TAG_NAME, "body").text
            if "Ships from and sold by" in body_text:
                raw = body_text.split("Ships from and sold by")[1].split("\n")[0]
                seller = raw.split(".")[0].strip()
                print(f"  ✅ [Amazon] Method 4 (ships+sold) found seller: {seller}")
            elif "Sold by" in body_text:
                after = body_text.split("Sold by")[1]
                lines = [l.strip() for l in after.split("\n") if l.strip()]
                for line in lines[:3]:
                    if line and line not in ("Payment", "Secure transaction", "Quantity:", "Add a Protection Plan:"):
                        seller = line
                        break
                print(f"  ✅ [Amazon] Method 4 (body scan) found seller: {seller}")
        except:
            seller = "Unknown"

    screenshot_name = f"Amazon_{sku}.png"
    driver.save_screenshot(screenshot_name)
    print(f"     [Amazon] SKU={sku} | Price={price} | Seller={seller}")

    return {
        "Seller_name": seller,
        "SKU": sku,
        "Region": "India",
        "Advertised_price": price,
        "Marketplace": "Amazon",
        "Screenshot": screenshot_name,
        "Live_link": url,
        "Last_time_collected": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }


# FLIPKART SCRAPER
def scrape_flipkart(url):
    driver.get(url)
    time.sleep(5)
    try:
        close_btn = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable(
                (By.XPATH, "//button[contains(text(),'✕')] | //span[contains(text(),'✕')]")
            )
        )
        close_btn.click()
        time.sleep(1)
    except:
        pass

    driver.execute_script("window.scrollBy(0, 600);")
    time.sleep(2)

    sku = url.split("pid=")[1].split("&")[0] if "pid=" in url else "Unknown"
    price = "N/A"
    seller = "Unknown"

    price_selectors = ["div.Nx9W0j", "div.UOCQB1", "div._30jeq3", "div._16Jk6d", "span.V_P69E"]
    for sel in price_selectors:
        try:
            elem = WebDriverWait(driver, 8).until(EC.visibility_of_element_located((By.CSS_SELECTOR, sel)))
            text = elem.text.strip()
            if "₹" in text:
                price = text
                break
        except:
            continue

    if price == "N/A":
        try:
            elems = driver.find_elements(By.XPATH, "//*[contains(text(), '₹')]")
            for e in elems:
                t = e.text.strip()
                if t.startswith("₹") and len(t) < 15:
                    price = t
                    break
        except:
            pass

    driver.execute_script("window.scrollBy(0, 800);")
    time.sleep(2)

    try:
        fulfilled_elems = driver.find_elements(
            By.XPATH, "//*[contains(text(), 'Fulfilled by')]"
        )
        for elem in fulfilled_elems:
            text = elem.text.strip()
            if "Fulfilled by" in text:
                raw = text.split("Fulfilled by")[1].strip()
                raw = raw.split("\n")[0].strip()
                import re
                raw = re.split(r'\d+\.\d+\s*★', raw)[0].strip()
                if raw:
                    seller = raw
                    break
    except:
        pass

    if seller == "Unknown":
        try:
            elem = driver.find_element(
                By.XPATH,
                "//*[starts-with(normalize-space(text()), 'Fulfilled by')]"
            )
            full = elem.text.strip()
            seller = full.replace("Fulfilled by", "").split("\n")[0].strip()
            import re
            seller = re.split(r'\d+\.\d+', seller)[0].strip()
        except:
            pass

    if seller == "Unknown":
        try:
            seller_label = driver.find_element(
                By.XPATH,
                "//*[normalize-space(text())='Seller' or normalize-space(text())='Seller:']/following-sibling::*[1]"
            )
            seller = seller_label.text.strip()
        except:
            pass

    if seller == "Unknown":
        try:
            body_text = driver.find_element(By.TAG_NAME, "body").text
            import re
            if "Fulfilled by" in body_text:
                raw = body_text.split("Fulfilled by")[1].split("\n")[0].strip()
                seller = re.split(r'\d+\.\d+', raw)[0].strip()
            elif "Seller:" in body_text:
                raw = body_text.split("Seller:")[1].split("\n")[0].strip()
                seller = raw
        except:
            seller = "Unknown"

    screenshot_name = f"Flipkart_{sku}.png"
    driver.save_screenshot(screenshot_name)
    print(f"     [Flipkart] SKU={sku} | Price={price} | Seller={seller}")

    return {
        "Seller_name": seller,
        "SKU": sku,
        "Region": "India",
        "Advertised_price": price,
        "Marketplace": "Flipkart",
        "Screenshot": screenshot_name,
        "Live_link": url,
        "Last_time_collected": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }


# EXECUTION the code
final_list = []

for item in products:
    print(f"\n{'='*55}")
    print(f"  Category: {item['Category']} | Brand: {item['Brand']}")
    print(f"{'='*55}")

    amz = scrape_amazon(item["Amazon"])
    if amz:
        final_list.append(amz)
    time.sleep(3)

    flp = scrape_flipkart(item["Flipkart"])
    if flp:
        final_list.append(flp)
    time.sleep(3)

# saving the results
df = pd.DataFrame(final_list)
output_file = "Seller_Table_Final_Corrected.csv"
df.to_csv(output_file, index=False)

print("\n\n--- FINAL RESULTS ---")
print(df[['Marketplace', 'SKU', 'Seller_name', 'Advertised_price']].to_string(index=False))
print(f"\nSaved to: {output_file}")

driver.quit()    
```
