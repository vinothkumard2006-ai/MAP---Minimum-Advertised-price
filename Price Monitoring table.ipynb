```python
import os
import re
from datetime import datetime, timedelta
from docx import Document
import pandas as pd



# 🔹 Folder to save letters
folder_name = "warning_letters2"
os.makedirs(folder_name, exist_ok=True)

# 🔹 Loop through each row
for index, row in df.iterrows():

    # ⏱ 24 hour deadline
    deadline = datetime.now() + timedelta(hours=24)
    deadline_str = deadline.strftime("%d-%m-%Y %I:%M %p")

    # 🔹 Letter content
    letter = f"""
Subject: LPP Violation Notice – Immediate Action Required

Dear Seller,

We have identified that your product listing is in violation of the pricing policy.

SKU: {row['SKU']}
Seller Name: {row['Seller_name']}

As per our policy, the listed price is lower than the allowed minimum price.

You are requested to correct the price before {deadline_str} (within 24 hours).

Failure to do so may result in actions such as:
* Product delisting
* Inventory blocking
* Further compliance measures

Kindly confirm once the correction is completed.

Regards,
Pricing Compliance Team
"""

    # 🔹 Clean file name (remove invalid characters)
    safe_sku = re.sub(r'[\\/*?:"<>|]', "", str(row['SKU']))
    safe_seller = re.sub(r'[\\/*?:"<>|]', "", str(row['Seller_name']))

    file_name = f"{safe_sku}{safe_seller}.docx".replace(" ", "")
    file_path = os.path.join(folder_name, file_name)

    # 🔹 Save as Word file
    doc = Document()
    doc.add_paragraph(letter)
    doc.save(file_path)

print("✅ LPP violation letters generated successfully!")
```
