```python
merge_df2 = pd.merge(Sellers_table,merge_df,on="SKU",how="inner")
merge_df2
```

```python
merge_df2.drop_duplicates(inplace=True)
```

```python
merge_df2
```

```python
merge_df2['adv_price'] = pd.to_numeric(merge_df2['adv_price'],errors='coerce')

conversion_rates = {
    'IND':1,
    'USA':91.17,
    'UK':91.07*1.27,
    'CAN':91.07*0.74,
    'UAE':91.07*0.27 }

merge_df2['adv_price_inr'] = merge_df2.apply(
    lambda row: row['adv_price']*conversion_rates.get(row['region'],1),axis=1)
merge_df2.head()
```

```python
merge_df2.isna().sum()
```

```python
merge_df2.shape
```

```python
violation = merge_df2[merge_df2['adv_price_inr'] < merge_df2['LPP']]
print("Total Violations:",violation['SKU'].count())
```

```python
violation['SKU'].nunique()
```

```python
violation.isna().sum()
```

```python
import os
import re

folder_name = "warning_letters"
os.makedirs(folder_name, exist_ok=True)

for index, row in violation.iterrows():

    # Clean seller name for file safety
    seller_clean = re.sub(r'[^a-zA-Z0-9]', '_', str(row['Seller_name']))
    
    file_name = f"{row['SKU']}_{seller_clean}.txt"
    file_path = os.path.join(folder_name, file_name)

    letter = f"""
Date: 2026-02-25

Subject: Price Violation Warning – LPP Policy

Dear {row['Seller_name']},

We have identified a pricing violation for the following product:

SKU              : {row['SKU']}
Marketplace      : {row['Marketplace']}
Region           : {row['region']}
Advertised Price : {row['adv_price_inr']}
LPP Price        : {row['LPP']}
Category         : {row['SUB_CATEGORY']}

Your advertised price is lower than the allowed LPP price.
This is a violation of our pricing policy.

Please correct the advertised price within 24 hours of receiving this notice.

Failure to comply within 24 hours may result in listening suspension or further action as per policy.

If you believe this notice was issued in error, please contact us immediately.

Regards,
MAP Compliance Team
"""

    with open(file_path, "w", encoding="utf-8") as f:
        f.write(letter)

print("All warning letters generated successfully!")
```
